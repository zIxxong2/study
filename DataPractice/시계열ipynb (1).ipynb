```python
value = pd.read_excel("/content/drive/MyDrive/동일건축 /1000(3).xlsx")
value
```

```python
date = value['Date']
date
```

```python
value = value.drop(['낙찰율','동일','Date'], axis=1)
value
```

```python
print("새로운 공고의 날짜를 입력해주세요: ")
input_date = input()
for col, item in value.iteritems():
  print("Name: ", col)
  #name = col.
  #print(item)

  value_company = pd.concat([date,item],axis=1, keys = ['ds','y'])
  #print(value_company)

  finalvalue_company = value_company[value_company.y != 0]
  finalvalue_company.columns = ['ds', 'y']
    
  m = Prophet(daily_seasonality=True)  
  m.fit(finalvalue_company)
    
  future = m.make_future_dataframe(periods=365)
  pred = m.predict(future)

  df_result = pd.DataFrame( columns=['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'trend_lower', 'trend_upper'])
  
  df1=[]
  result = pred.loc[pred['ds'] == input_date]
  #df1['name']= name
  df1 = pd.DataFrame(result[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'trend_lower', 'trend_upper']])
  
  df_result =pd.concat([df_result,df1])
  print(df_result)


```
